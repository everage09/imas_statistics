```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

from urllib.request import urlopen
import urllib.request
import time
import os

import pandas as pd
import numpy as np
```

```python
MAX_WAIT_TIME = 3
idol_arr = np.zeros((52,18))
idol_df = pd.DataFrame(data=idol_arr, 
                       columns=["아이돌넘버", "이름", "한자", "히라가나", "나이", "키", "몸무게", "B-W-H", "생일", "별자리"
                               , "혈액형", "자주 쓰는 손", "출신지", "취미", "특기", "이미지 컬러 코드", "성우", "타입"])
idol_df
```

```python
allstar = ['아마미 하루카', '키사라기 치하야','호시이 미키', '하기와라 유키호', '타카츠키 야요이', '키쿠치 마코토',
           '미나세 이오리', '시죠 타카네', '아키즈키 리츠코', '미우라 아즈사', '후타미 아미', '후타미 마미',
          '가나하 히비키']
princess = ['카스가 미라이','타나카 코토하','사타케 미나코','토쿠가와 마츠리','나나오 유리코','타카야마 사요코',
           '마츠다 아리사','코사카 우미','나카타니 이쿠','에밀리 스튜어트','야부키 카나','요코야마 나오','후쿠다 노리코']
fairy = ['모가미 시즈카',' 토코로 메구미','로코(아이돌 마스터 밀리언 라이브)','텐쿠바시 토모카','키타자와 시호','마이하마 아유무','니카이도 치즈루',
        '마카베 미즈키','모모세 리오','나가요시 스바루','스오 모모코','줄리아(아이돌 마스터 밀리언 라이브)','시라이시 츠무기']
angel = ['이부키 츠바사','시마바라 엘레나','하코자키 세리카','노노하라 아카네','모치즈키 안나','키노시타 히나타','바바 코노미',
        '오오가미 타마키','토요카와 후우카','미야오 미야','시노미야 카렌','키타카미 레이카','사쿠라모리 카오리']
```

```python
# setting urllib opener 
opener = urllib.request.build_opener()
opener.addheaders = [('User-Agent', 'Mozila/5.0')]
urllib.request.install_opener(opener)

# initialize Google Chrome webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36")
```

```python
def get_driver():
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    driver.maximize_window()
    driver.implicitly_wait(MAX_WAIT_TIME)
    return driver
```

```python
for i, idol in enumerate(allstar):
    driver = get_driver()
    # search for idol page in wiki
    try:
        driver.get("https://namu.wiki/w/"+idol)
        time.sleep(2)
    except:
        print(f"Failed to open url: 'https://namu.wiki/w/{idol}', check name:{idol}'")
        continue
    
    # retrieve profile table.
    try:
        profile = driver.find_element(By.XPATH, 
                                      '//*[@id="lYnqWtHhT"]/div[2]/div/div/div/div/div/div/div/div/div[1]/div[7]/div/div/div/div/div/div/div/div/div[5]/div/div/div/div/div[17]/div[1]/div/div[4]/div[1]/div/table')
    except:
        try:
            profile = driver.find_element(By.XPATH, 
                                      '//*[@id="lYnqWtHhT"]/div[2]/div/div/div/div/div/div/div/div/div[1]/div[7]/div/div/div/div/div/div/div/div/div[5]/div/div/div/div/div[17]/div[1]/div/div[5]/div[1]/div/table')
        except:
            profile = driver.find_element(By.XPATH, '//*[@id="lYnqWtHhT"]/div[2]/div/div/div/div/div/div/div/div/div[1]/div[7]/div/div/div/div/div/div/div/div/div[5]/div/div/div/div/div[17]/div[1]/div/div[7]/div[1]/div/table')
    tr = profile.find_elements(By.CSS_SELECTOR, "tr")
    
    # extract name
    name = tr[0].find_elements(By.CSS_SELECTOR, "ruby")
    kanji = ""
    hiragana = ""
    for name_part in name:
        k, h = name_part.text.split('\n')
        kanji += k
        kanji += " "
        hiragana += h
        hiragana += " "
    kanji = kanji.rstrip()
    hiragana = hiragana.rstrip()
    #print("이름:")
    #print(kanji)
    #print(hiragana)
    #print("__________")
    idol_df.loc[i, "아이돌넘버"] = i+1
    idol_df.loc[i, "이름":"히라가나"] = [idol, kanji, hiragana]
    
    # extract other information -> tr[1] contains idol image
    for j in range(2,len(tr)):
        # extract profile
        category, data = tr[j].find_elements(By.CSS_SELECTOR, "td")
        try:
            info = data.text.replace('\n', ' ').rstrip("[1]")
        except:
            pass
        
        try:
            if (category.text == "이미지 컬러"):
                idol_df.loc[i, "이미지 컬러 코드"] = info
            elif (category.text == "쓰리 사이즈"):
                idol_df.loc[i, "B-W-H"] = info
            else:
                idol_df.loc[i, category.text] = info
        except:
            print(f"{category.text}:{info} - failed")
            pass
        #print(f"{category.text}: {info}")
    driver.close()
```
